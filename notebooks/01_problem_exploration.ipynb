# 01 - Problem Exploration

This notebook introduces end-of-turn detection as a turn-taking problem for conversational voice agents. The goal is to inspect the synthetic dataset and identify signals that may help decide whether a speaker is finished.

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
data = pd.read_csv(PROJECT_ROOT / 'data' / 'sample_turns.csv')
data.head()
```

```python
data['label_end_of_turn'].value_counts(normalize=True).rename('proportion')
```

```python
data.groupby('label_end_of_turn')[['pause_duration_ms', 'speech_duration_ms', 'num_words', 'syntactic_completeness_score']].mean().round(2)
```

Initial expectation: completed turns should often have longer pauses and higher completeness scores, but the ambiguous cases are the interesting part. Backchannels may be short but complete, while incomplete sentences may have a pause without being complete.
