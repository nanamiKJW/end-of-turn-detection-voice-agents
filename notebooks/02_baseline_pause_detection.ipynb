# 02 - Baseline Pause Detection

A silence threshold is the simplest end-of-turn detector. This notebook compares several pause thresholds and motivates why a text-aware model can be useful.

```python
from pathlib import Path
import sys
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

from baseline_pause_detector import evaluate_thresholds

data = pd.read_csv(PROJECT_ROOT / 'data' / 'sample_turns.csv')
evaluate_thresholds(data).round(3)
```

```python
threshold = 700
errors = data.assign(pred=(data['pause_duration_ms'] >= threshold).astype(int))
errors.loc[errors['pred'] != errors['label_end_of_turn'], ['utterance_text', 'pause_duration_ms', 'label_end_of_turn', 'pred']]
```

The baseline is interpretable, but it treats all silences the same. In a real system, a user may pause to think, search for a word, or self-correct.
